```python
import math

# Initialize the board
board = [' ' for _ in range(9)]

def print_board():
    for i in range(0, 9, 3):
        print(f" {board[i]} | {board[i+1]} | {board[i+2]} ")
        if i < 6:
            print("---+---+---")
    print()

def is_winner(b, player):
    # All possible winning combinations
    win_states = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]              # Diagonals
    ]
    return any(all(b[cell] == player for cell in state) for state in win_states)

def is_board_full(b):
    return ' ' not in b

def minimax(b, depth, is_maximizing):
    if is_winner(b, 'O'):
        return 1
    if is_winner(b, 'X'):
        return -1
    if is_board_full(b):
        return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'O'
                score = minimax(b, depth + 1, False)
                b[i] = ' '
                best_score = max(best_score, score)
        return best_score
    else:
        best_score = math.inf
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'X'
                score = minimax(b, depth + 1, True)
                b[i] = ' '
                best_score = min(best_score, score)
        return best_score

def get_best_move():
    best_score = -math.inf
    move = -1

    for i in range(9):
        if board[i] == ' ':
            board[i] = 'O'
            score = minimax(board, 0, False)
            board[i] = ' '
            if score > best_score:
                best_score = score
                move = i

    return move

# Main Game Loop
print("Welcome to Tic-Tac-Toe!")
print("Board positions are:")
print("0 | 1 | 2")
print("--+---+--")
print("3 | 4 | 5")
print("--+---+--")
print("6 | 7 | 8\n")

print_board()

while True:
    # Human Turn
    try:
        move = int(input("Enter your move (0-8): "))
        if move < 0 or move > 8:
            print("Please enter a number between 0 and 8.")
            continue
        if board[move] != ' ':
            print("Slot already filled! Try again.")
            continue
    except ValueError:
        print("Invalid input. Enter a valid number.")
        continue

    board[move] = 'X'
    print_board()

    if is_winner(board, 'X'):
        print("🎉 You win!")
        break

    if is_board_full(board):
        print("It's a tie!")
        break

    # AI Turn
    print("AI is thinking...")
    ai_move = get_best_move()
    board[ai_move] = 'O'
    print_board()

    if is_winner(board, 'O'):
        print("🤖 AI wins!")
        break

    if is_board_full(board):
        print("It's a tie!")
        break
```
