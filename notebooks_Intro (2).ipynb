```python
import re

def rule_based_chatbot(user_input):
    # Normalize input to lowercase
    user_input = user_input.lower().strip()
    
    # Predefined rules using simple pattern matching
    if re.search(r'\b(hello|hi|hey|hola)\b', user_input):
        return "Hello! How can I help you today?"
        
    elif re.search(r'\b(how are you|how\'s it going)\b', user_input):
        return "I am doing well, thank you for asking! How are you?"
        
    elif re.search(r'\b(name|who are you)\b', user_input):
        return "I am a simple rule-based AI chatbot built to assist you."
        
    elif re.search(r'\b(help|commands|what can you do)\b', user_input):
        return "I can answer basic greetings, tell you my name, or help you exit the chat."
        
    elif re.search(r'\b(bye|exit|quit|goodbye)\b', user_input):
        return "Goodbye! Have a great day!"
        
    else:
        return "I'm sorry, I didn't quite catch that. Could you please rephrase?"

# Simulation loop
print("Chatbot: Hello! Type 'bye' to exit.")
while True:
    user_text = input("You: ")
    response = rule_based_chatbot(user_text)
    print(f"Chatbot: {response}")
    if response == "Goodbye! Have a great day!":
        break
```
